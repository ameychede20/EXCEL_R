```python
import pandas as pd

# Load the datasets
movies_df = pd.read_csv('/content/my_movies.csv')
books_df = pd.read_csv('/content/book.csv')
movies_onehot = movies_df.drop(columns=['V1', 'V2', 'V3', 'V4', 'V5'])
from mlxtend.frequent_patterns import apriori, association_rules

def apply_apriori(df, support, confidence):
    frequent_itemsets = apriori(df, min_support=support, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=confidence)
    return rules
support_values = [0.2, 0.15, 0.1]
confidence_values = [0.7, 0.6, 0.5]

for support in support_values:
    for confidence in confidence_values:
        rules = apply_apriori(movies_onehot, support, confidence)
        print(f"Support: {support}, Confidence: {confidence}, Rules Generated: {len(rules)}")
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming `rules` is your DataFrame of association rules for one of the experiments
plt.figure(figsize=(10, 6))
sns.scatterplot(x='support', y='confidence', data=rules, size='lift', hue='lift', palette='viridis')
plt.title('Support vs Confidence')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```
