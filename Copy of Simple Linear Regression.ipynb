 Predicting Delivery Time based on Sorting Time

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load Delivery Time data
delivery_time_data = pd.read_csv('/content/delivery_time.csv')

# EDA: Scatter plot for Delivery Time vs Sorting Time
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sorting Time', y='Delivery Time', data=delivery_time_data)
plt.title('Delivery Time vs Sorting Time')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.show()

# Prepare the data
X_delivery = delivery_time_data[['Sorting Time']]
y_delivery = delivery_time_data['Delivery Time']

# Split data into training and testing sets
X_delivery_train, X_delivery_test, y_delivery_train, y_delivery_test = train_test_split(X_delivery, y_delivery, test_size=0.2, random_state=42)

# Build Linear Regression model
delivery_model = LinearRegression()
delivery_model.fit(X_delivery_train, y_delivery_train)

# Predict and evaluate the model
y_delivery_pred = delivery_model.predict(X_delivery_test)
delivery_mse = mean_squared_error(y_delivery_test, y_delivery_pred)
delivery_r2 = r2_score(y_delivery_test, y_delivery_pred)

print(f"Delivery Model MSE: {delivery_mse}, R2: {delivery_r2}")
```

Predicting Salary based on Years of Experience

```python
salary_data = pd.read_csv('/content/Salary_Data.csv')

# EDA: Scatter plot for Salary vs Years of Experience
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearsExperience', y='Salary', data=salary_data)
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

# Prepare the data
X_salary = salary_data[['YearsExperience']]
y_salary = salary_data['Salary']

# Split data into training and testing sets
X_salary_train, X_salary_test, y_salary_train, y_salary_test = train_test_split(X_salary, y_salary, test_size=0.2, random_state=42)

# Build Linear Regression model
salary_model = LinearRegression()
salary_model.fit(X_salary_train, y_salary_train)

# Predict and evaluate the model
y_salary_pred = salary_model.predict(X_salary_test)
salary_mse = mean_squared_error(y_salary_test, y_salary_pred)
salary_r2 = r2_score(y_salary_test, y_salary_pred)

print(f"Salary Model MSE: {salary_mse}, R2: {salary_r2}")
```
